```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blink_img = np.zeros(shape = (512, 512, 3), dtype = np.int16)
```

```python
blink_img.shape
```

```python
plt.imshow(blink_img)
```

```python
cv2.rectangle(blink_img, pt1=(484,0), pt2=(510,150), color = (0,255,0), thickness = 10)
```

```python
plt.imshow(blink_img)
```

```python
cv2.circle(img = blink_img, center = (100,100),radius = 50, color = (255,0,0), thickness = 8)
plt.imshow(blink_img)
```

```python
# adding fonts and pollygons

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blink_img, text = 'hello', org = (10, 500),fontFace = font, fontScale = 4, color = (255,255,255), thickness = 3, lineType = cv2.LINE_AA)

plt.imshow(blink_img)
```

```python
# own custom pollygons

blank_img = np.zeros(shape = (512, 512, 3), dtype = np.int16)
```

```python
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```
