```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# matplotlib ----> RGB RED, GREEN, BLUE
# OPENCV ----> BLUE, GREEN, RED


# now we need to convert the image to RGB matplotlib version
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img) 

# after converting we can see the real image
```

```python
img_gray = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE) 
```

```python
# importing image in grayscale format when we load the image

img_gray.shape
```

```python
plt.imshow(img_gray) # this happen because the default color mapping in opencv
```

```python
# after color mapping cmap = colormapping

plt.imshow(img_gray,cmap ='gray')
```

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img,(1000,400)) # resizing (lenght, width)
```

```python
plt.imshow(new_img)
```

```python
# resizing using ratio 
# in this we can use width ratio and height ratio
```

```python
w_ratio = 0.5
h_ratio = 0.5

# we can change this as a ratio of our own 
# as our own ratio
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
# we can flip the image using flip function

new_img = cv2.flip(fix_img,0)
plt.imshow(new_img)
```
