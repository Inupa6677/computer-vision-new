```python
img = cv2.imread('Computer-Vision-with-Python/DATA/rainbow.jpg', 0)
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
img.max() # for this image max val is 214 but possible max is 214
```

```python
ret, thresh1 = cv2.threshold(img,127,255, cv2.THRESH_BINARY)

# (image, threshold, max value)
```

```python
ret
```

```python
plt.imshow(thresh1, cmap='gray')

# this is a binary image all the value is 0 or 255
```

```python
ret, thresh1 = cv2.threshold(img,127,255, cv2.THRESH_BINARY_INV)

# binary inverse of the image
```

```python
plt.imshow(thresh1, cmap='gray')
```
